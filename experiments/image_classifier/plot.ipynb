```python
import torch
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib as mpl

mpl.rcParams['figure.dpi'] = 200
```

Esta función sirve para graficar el error de entrenamiento para un único learning rate, imprimiendo el error en cada iteración de la validación cruzada

```python
def loss_plot(losses, labels=None, legend_title=None):
    font_size = 15
    print(losses.shape)

    plt.figure(figsize=(3,3))
    sns.set_style("darkgrid",{"axes.edgecolor": ".1"})

    for fold in losses:
        sns.lineplot(data=fold)

    plt.yscale('log')
    plt.ylabel("Error", fontsize=font_size)
    plt.xlabel("Época", fontsize=font_size)

    if labels is not None:
        lgd = plt.legend(labels=labels, loc='center left', bbox_to_anchor=(1.1,0.5), fancybox=True, ncol=3, title=legend_title)
    plt.show()
```
